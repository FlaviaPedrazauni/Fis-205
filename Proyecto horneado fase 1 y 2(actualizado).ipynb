```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve_banded
```

Iniciamos estableciendo los parámetros geométricos y temporales, las propiedades termofísicas de las masa, las cuales en este caso, serán usadas como constanstes en primera instancia, simplificando los cálculos del modelo como "fase 1", tambien tenemos las condiciones iniciales y de horneado, donde la variable H siendo el coeficiente dado por el tipo de transmisión de calor, será un valor estandar de un horno con radiación y convección natural.

```python
#Parámetros geométricos y de malla
L = 0.04             # Espesor desde el centro a la superficie [m]
Nx = 140                  # Número de nodos espaciales
dx = L / (Nx - 1)         # Paso espacial [m]
X = np.linspace(0, L, Nx) # Posiciones [m]
# Parámetros temporales
t_max = 1800          # Tiempo máixmo de simulación aprox 30 min de horneado (s)
dt = 0.5                  # Paso de tiempo (s) [Pequeño para mantener la estabilidad del método C-N]
Nt = int(t_max / dt)      # Número de pasos temporales
# Propiedades termofísicas de la masa (constantes)
rho = 800           # densidad [kg/m³]
Cp  =2000           # calor específico [J/(kg·K)]
k   = 0.5           # conductividad térmica [W/(m·K)]
alpha = k / (rho * Cp)  # difusividad térmica [m²/s]  ≈ 3.125e-7
# Condiciones de horneado e iniciales 
T_horno = 180.0          # Temperatura del horno (°C)
H = 25.0            # Coeficiente convención + radiación [W/(m²·K)] 
T_ini = 25.0         # Temperatura inicial en todo el dominio [°C]
W_ini = 0.60          # Humedad inicial base seca [kg agua/kg sólido]
gamma = (H* dx) / k # Parámetro para la condición de Robin 
# Coeficiente de estabilidad para Crank-Nicolson 
r = (alpha * dt) / (2 * dx**2)
```

**Resolución Númerica del modelo**


```python
# Construcción de la matriz diagonal 
A = np.zeros((3, Nx))
A[1, :] = 1 + 2*r     # Diagonal principal (nodos internos)
A[0, 1:] = -r         # Diagonal superior
A[2, :-1] = -r        # Diagonal inferior

# Ajuste por simetría en el centro (x=0) 
A[1, 0] = 1 + 2*r
A[0, 1] = -2*r 

# Ajuste por condición de Robin en la superficie (x=L) 
A[1, -1] = 1 + 2*r * (1 + gamma)
A[2, -2] = -2*r

# BUCLE TEMPORAL 
T = np.full(Nx, 25.0) # Temperatura inicial (ambiente)
temp_centro = []    # La evolución en x = 0(centro)
temp_superficie = [] # La evolución en x = L(superficie)

for n in range(Nt):
    # Construimos el vector B 
    B = np.zeros(Nx)
    
    # Nodos internos
    for i in range(1, Nx-1):
        B[i] = r*T[i-1] + (1-2*r)*T[i] + r*T[i+1]
        
    # Frontera Centro (Simetría)
    B[0] = (1-2*r)*T[0] + 2*r*T[1]
    
    # Frontera Superficie (Robin)
    B[-1] = 2*r*T[-2] + (1 - 2*r*(1 + gamma))*T[-1] + 4*r*gamma*T_horno
    
    # Resolvemos el sistema lineal
    T = solve_banded((1, 1), A, B)
    # Guardamos la evolución temporal
    temp_centro.append(T[0])
    temp_superficie.append(T[-1])

print(f"Temperatura final de la superficie :{temp_superficie[-1]:.0f}°C")
print(f"Temperatura final del centro :{temp_centro[-1]:.0f}°C")
    
```

**Visualización de la resolución del modelo:**

```python
# Visualización
tiempo_eje = np.linspace(0, t_max/60, Nt) # Eje de tiempo en minutos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico A: Perfil de temperatura final (Posición)
x_cm = np.linspace(0, L, Nx) * 100
ax1.plot(x_cm, T, color='skyblue', lw=2, label='Tiempo total de horneado (30 min)')
ax1.axhline(100, color='red', linestyle='--', alpha=0.6, label='Zona Evaporación')
ax1.set_xlabel('Posición desde el centro (cm)')
ax1.set_ylabel('Temperatura (°C)')
ax1.set_title('Distribución Espacial de Temperatura')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Gráfico B: Evolución Temporal (Centro vs Borde)
ax2.plot(tiempo_eje, temp_centro, label='Centro (0 cm)', color='yellow')
ax2.plot(tiempo_eje, temp_superficie, label='Superficie (5 cm)', color='fuchsia')
ax2.axhline(100, color='red', linestyle='--', alpha=0.6)
ax2.set_xlabel('Tiempo (minutos)')
ax2.set_ylabel('Temperatura (°C)')
ax2.set_title('Evolución Temporal de la Temperatura en el Centro y Superficie')
ax2.grid(True, alpha=0.3)
ax2.legend()
plt.tight_layout()
plt.show()
```

A partir de los resultados anteriores y comparando con los resultados de autores como Zurcher, el cual utiliza realiza una aproximación con varibles constantes como lo realizamos en este caso, donde ambos obtenemos que en la superficie del pan la temperatura se eleva rapidamente, indicando que dentro de la masa de pan hay una evaporación del agua y a su vez una formación de corteza en esta superfice, pasada la temperatura de evaporación(100°C T en la que se evapora el agua). Aún asi nos topamos con que en el centro apenas logra llegar a los 59°C, esto nos exige extender el tiempo de medición, en donde recien a los 60 minutos llega a una temperatura de 99°C, lo que implica que esta iniciando el proceso de evaporación del agua en esa zona y cocción. 
Tomando en cuenta lo anterior y a otros autores, tales como Zheleva y Purlis, continuamos a la fase 2 donde modelaremos el problema como un cilindro infinito y con propiedades que dependen de la temperatura y la humedad de la masa, junto a un factor de evaporación, esperando asi obtener una meseta a los 100°C para el centro de la masa, donde ocurrira el proceso de cocción y evaporación del agua por el flujo de este calor, solo formando la corteza en la superficie como obtuvimos en la fase 1 y manteniendo un mínimo de humedad 

**Fase 2: Propiedades Variables y término de evaporación**

```python
# --- Geometria cilindrica 1D ---
R   = 0.04                 # radio del cilindro [m] 
Nx  = 140                  # numero de nodos en r
dr  = R / (Nx - 1)         # paso espacial [m]
rg  = np.linspace(0, R, Nx)  # coordenada radial [m]: rg[0]=0 (centro), rg[-1]=R (superficie)

# --- Tiempo ---
t_max = 1800               # tiempo total [s] (30 min)
dt    = 0.5          # paso temporal [s]
Nt    = int(t_max / dt)

# --- Condiciones de horneado e iniciales ---
T_horno = 180.0            # temperatura del horno [C]
H       = 25.0             # coef. convectivo+radiativo global [W/(m2 K)]
T_ini   = 25.0             # temperatura inicial uniforme [C]
W_ini   = 0.60             # humedad inicial base seca [kg agua/kg solido] 

# --- Constantes del cambio de fase ---
Tf  = 373.16               # temperatura de evaporacion [K] (100 C)
dTb = 0.5                  # semiancho de la banda de cambio de fase [K]
Lv  = 2.3e6                # calor latente de evaporacion [J/kg]
rho_s = 241.76             # densidad del solido seco [kg/m3] 

# Conversion de la temperatura a kelvin para los cálculos
def C2K(Tc): return Tc + 273.16
def K2C(Tk): return Tk - 273.16
```

**Definición de las propiedades variables**
-Notar que se define una función `smoothH`, la cual es una Heaviside suavizada (≈0 debajo de 100 °C, ≈1 arriba) que usamos para mezclar las ramas miga/corteza sin saltos, en la resolución númerica.
-Las Propiedades variables son obtenidas por las referencias [Purlis]

```python
def smoothH(T):
    """Heaviside suavizada centrada en Tf. ~0 (miga) debajo, ~1 (corteza) arriba."""
    return 1.0 / (1.0 + np.exp(-(T - Tf) / (dTb / 3.0)))

def k_eff(T):
    """Conductividad termica efectiva [W/m K], T en K."""
    k_miga = 0.9 / (1.0 + np.exp(-0.1 * (T - 353.16))) + 0.2
    s = smoothH(T)
    return (1.0 - s) * k_miga + s * 0.2

def Cp_sens(T, W):
    """Cp sensible (sin latente) [J/kg K], T en K """
    Cps = 5.0 * T + 25.0                                        # solido seco
    Cpw = (5.207 - 73.17e-4 * T + 1.35e-5 * T**2) * 1000.0      # agua
    return Cps + W * Cpw

def rho_eff(T):
    """Densidad efectiva [kg/m3], T en K"""
    s = smoothH(T)
    return (1.0 - s) * 321.31 + s * 180.61
"""
# --- Verificacion: como varian con T ---
print('T(C) |  k(W/mK) | Cp*(J/kgK) | rho(kg/m3)')
print('-'*48)
for Tc in [25, 60, 90, 100, 140, 180]:
    Tk = C2K(Tc)
    print(f' {Tc:4.0f} |  {k_eff(Tk):.3f}   |   {Cp_sens(Tk,W_ini):6.0f}   |   {rho_eff(Tk):6.1f}')
"""
```

**Resolución con Crank-Nicolson cilíndrico:**
Definimos la función que discretiza $\rho C_p\,\partial T/\partial t = \frac{1}{r}\partial_r(r\,k\,\partial_r T)$. 
Donde el interior se abarca con k y r en las interfaces, en el centro se utiliza L'Hôpital (factor 2) para evitar un factor que tiende a infinito en ciertas condiciones y en la superficie usamos condición de Robin. 

```python
def construc_sis(T, W):
    N=len(T); k=k_eff(T); cp=Cp_sens(T,W); rh=rho_eff(T)
    A=np.zeros((3,N)); B=np.zeros(N)
    for i in range(1, N-1):
        ri=rg[i]; r_p=0.5*(rg[i]+rg[i+1]); r_m=0.5*(rg[i]+rg[i-1])
        k_p=0.5*(k[i]+k[i+1]); k_m=0.5*(k[i]+k[i-1])
        coef=dt/(rh[i]*cp[i]*ri*dr*dr); cp_=0.5*coef*r_p*k_p; cm_=0.5*coef*r_m*k_m
        A[2,i-1]=-cm_; A[1,i]=1+cp_+cm_; A[0,i+1]=-cp_
        B[i]=cm_*T[i-1]+(1-cp_-cm_)*T[i]+cp_*T[i+1]
    c0=2.0*dt*k[0]/(rh[0]*cp[0]*dr*dr); A[1,0]=1+2*c0; A[0,1]=-2*c0; B[0]=(1-2*c0)*T[0]+2*c0*T[1]
    ks=k[-1]; gamma=H*dr/ks; rN=0.5*dt*ks/(rh[-1]*cp[-1]*dr*dr)
    A[2,-2]=-2*rN; A[1,-1]=1+2*rN*(1+gamma)
    B[-1]=2*rN*T[-2]+(1-2*rN*(1+gamma))*T[-1]+4*rN*gamma*C2K(T_horno)
    return A, B
```

**Factor S que representa la evaporación del agua del sistema:**
Este se define en una función para luego conbinarlo con la ecuación de calor discretizada

```python
def factor_evaporacion(T_arr, W_arr):
    """Evaporacion por operator-splitting"""
    for i in range(len(T_arr)):
        if T_arr[i] > Tf and W_arr[i] > 0.0:
            rh = rho_eff(T_arr[i])
            cp = Cp_sens(T_arr[i], W_arr[i])
            E_exceso = rh * cp * (T_arr[i] - Tf) * dr      # energia por encima de 100 C
            E_max    = W_arr[i] * rho_s * Lv * dr           # energia para evaporar TODA el agua
            if E_exceso <= E_max:
                dW = E_exceso / (rho_s * Lv * dr)           # agua evaporada 
                W_arr[i] = W_arr[i] - dW
                T_arr[i] = Tf                               # queda en la meseta de 100 C
            else:
                W_arr[i] = 0.0
                E_resto  = E_exceso - E_max
                T_arr[i] = Tf + E_resto / (rh * cp * dr)    # se seco -> recalienta
    return T_arr, W_arr
```

**Bucle principal:**
-La metodología es resolver la ecuación de calor con la función construc_sis(T,W), luego a lo resultante se le aplica el factor S de evaporación, el cual ayuda a visualizar mejor la perdida de peso en el proceso de horneado.
-Se utiliza un arreglo en el paso temporal, para obtener un factor de fourier cercano a 1,1, el cual estabiliza oscilación generada por el operator-splitting y a su vez mejora las curvas de tempertura gráficadas.

```python
T = np.full(Nx, C2K(T_ini)); W = np.full(Nx, W_ini)
w_cell = rg.copy(); w_cell[0] = dr/8.0
m_water0 = np.sum(W*w_cell); m_solid = np.sum(1.0*w_cell)
save = max(1, Nt//60)
tmin=[0.0]; Tc=[T_ini]; Ts=[T_ini]; Wsf=[W_ini]; WL=[0.0]
n_sub = 3
dt_sub = dt / n_sub

for n in range(Nt):
    for _ in range(n_sub):
        A, B = construc_sis(T, W)   
        T = solve_banded((1, 1), A, B)
        T, W = factor_evaporacion(T, W)     
    if (n+1) % save == 0:
        mw = np.sum(W*w_cell)
        tmin.append((n+1)*dt/60); Tc.append(K2C(T[0])); Ts.append(K2C(T[-1]))
        Wsf.append(W[-1]); WL.append(100*(m_water0 - mw)/(m_water0 + m_solid))

print(f'T centro final = {Tc[-1]:.1f} C | T superficie final = {Ts[-1]:.1f} C | perdida de peso = {WL[-1]:.1f} %')
```

**Visualización de los resultados:**
-Tenemos 3 gráficos:
1)Temperatura en función del tiempo de horneado, en donde se visualiza como aumenta la temperatura del centro del pan y de la superficie.
2)Porcentaje de perdida de peso en función del tiempo de horneado, en donde se visualiza como a medida que avanza el tiempo de horneado la perdida de peso va aumentando. 
3)Humedad final del pan en función del radio de este, en donde se visualiza como a medida que nos acercamos a la superficie, la humedad decae hasta secarse.

```python
ig,ax = plt.subplots(1, 3, figsize=(16,4.8))
#Grafico 1:Temperatura en función del tiempo de horneado
ax[0].plot(tmin, Tc, 'gold', lw=2.4, label='Centro')
ax[0].plot(tmin, Ts, 'coral', lw=2.4, label='Superficie')
ax[0].axhline(100, color='gray', ls='--', lw=1)
ax[0].set_xlabel('Tiempo [min]'); ax[0].set_ylabel('T [C]'); ax[0].set_title('Temperaturas')
ax[0].legend(); ax[0].grid(alpha=0.3)
#Grafico 2:Porcentaje de perdida de peso en función del tiempo de horneado
ax[1].plot(tmin, WL, 'magenta', lw=2.4)
ax[1].set_xlabel('Tiempo [min]'); ax[1].set_ylabel('Perdida de peso [%]'); ax[1].set_title('Perdida de peso'); ax[1].grid(alpha=0.3)
#Grafico 3:Humedad final del pan en función del radio de este
ax[2].plot(rg*100, W, 'royalblue', lw=2.4)
ax[2].set_xlabel('r [cm]'); ax[2].set_ylabel('W [kg/kg]'); ax[2].set_title('Humedad final (frente de secado)')
ax[2].set_ylim(-0.03, W_ini*1.1); ax[2].grid(alpha=0.3)
plt.tight_layout(); plt.show()
```
